```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("penguins")
df
```

```python
bar_order = ["Dream", "Torgersen", "Biscoe"]

sns.barplot(
    x="island", 
    y="bill_length_mm", 
    data=df, 
    hue="sex", 
    order=bar_order, 
    hue_order=["Female", "Male"],
    errorbar=('ci', 1)
)

plt.show()
```

```python
# Changing the order of the bar

bar_order = ["Dream", "Torgersen", "Biscoe"]

sns.barplot(
    x="island", 
    y="bill_length_mm", 
    data=df, 
    hue="sex", 
    order=bar_order, 
    hue_order=["Female", "Male"],
    errorbar=('ci', 1)
)

plt.show()
```

```python
# For horizenttal bar

sns.barplot(
      x="bill_depth_mm", 
      y="bill_length_mm", 
      data=df,
      orient="h"
)

plt.show()
```

```python
# For colors

sns.barplot(
      x="island", 
      y="bill_length_mm", 
      data=df,
      hue="sex",
      palette="turbo_r",
)

plt.show()
```

```python
# saturation of the bars/color of the error

sns.barplot(
      x="island", 
      y="bill_length_mm", 
      data=df,
      hue="sex",
      palette="turbo_r",
      saturation=0.5,
      errcolor='b'
)

plt.show()
```
