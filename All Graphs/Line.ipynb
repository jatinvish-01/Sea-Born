```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var = [1,2,3,4,5,6,7]
var_2 = [14,23,65,76,32,15,44]

x = pd.DataFrame({
      "var":var,
      "var_2":var_2
})

sns.lineplot(x="var", y="var_2", data=x)

plt.show()
```
